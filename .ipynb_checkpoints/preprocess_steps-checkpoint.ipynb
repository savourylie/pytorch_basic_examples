```python
import numpy as np
import pandas as pd

data_folder  = '../data/titanic/raw/'
```

```python
data = pd.read_csv(data_folder + 'train.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
def drop_unwatned_cols(data, cols=[]):
    return data.drop(cols, axis=1)
```

```python
data = drop_unwatned_cols(data, cols=['Cabin', 'PassengerId', 'Name', 'Ticket'])
```

```python
data['Age_na'] = 0
```

```python
data.ix[data['Age'].isnull(), 'Age_na'] = 1
```

```python
data.head()
```

```python
data['Sex'].value_counts()
```

```python
data['Embarked'].value_counts()
```

```python
data = pd.get_dummies(data, columns=['Sex', 'Embarked', 'Pclass'], drop_first=True)
```

```python
data.head()
```

```python
data = data.fillna(value={'Age': data['Age'].mean()})
```

```python
def min_max_normalize(data, cols=[]):
    data = data.copy()
    for colname in cols:
        data[colname] = (data[colname] - data[colname].min()) / (data[colname].max() - data[colname].min())
        
    return data
```

```python
data = min_max_normalize(data, cols=['Age', 'Fare'])
```

```python
data.describe()
```

```python
features, label = data.drop(['Survived'], axis=1), data['Survived']
```

```python
features.head()
```

```python
label.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=1113)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(512, random_state=1113)
rf.fit(X_train, y_train)

y_pred = rf.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test, y_pred[:, 1]))
```

```python
X_train.to_csv('X_train.csv')
X_test.to_csv('X_test.csv')
y_train.to_csv('y_train.csv')
y_test.to_csv('y_test.csv')
```
